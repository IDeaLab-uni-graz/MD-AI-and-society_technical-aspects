## Zellen erstellen, editieren und ausführen

Das Notebook besteht aus einzelnen **Zellen** die als Inhalt Text oder Code haben können (dazu später mehr).


### Modaler Editor

Das Jupyter Notebook hat eine **modale** Nutzer:innenoberfläche. Das bedeutet, dass Tastatureingaben unterschiedliche Aktionen bewirken, je nachdem in welchem **Modus** sich das Notebook gerade befindet. Es gibt zwei Modi:
1. der Editier-Modus
2. der Kommando-Modus

Im **Editier-Modus** wird in Google Colab oberhalb der Zelle eine Reihe von Formatierungsoptionen für den Text angezeigt. Im **Kommando-Modus** verschwinden die Formatierungsoptionen.

<font color='green'>Aktivieren Sie den <b>Editier-Modus</b> indem Sie</font>
<ul class="outside">
<li><font color='green'><b>enter</b> drücken während sich der Mauszeiger auf der Zelle befindet oder</font></li>
<li><font color='green'>mit der Maus in das Innere einer Zelle doppelklicken oder</font></li>
<li><font color='green'>im Menü der Zelle auf <b>Bearbeiten</b> klicken.</font></li>
</ul>

<font color='green'>Aktivieren Sie den <b>Kommando-Modus</b>, indem Sie</font>
<ul class="outside">
<li><font color='green'><b>esc</b> drücken während sich der Mauszeiger auf der Zelle befindet oder</font></li>
<li><font color='green'>mit der Maus links außerhalb der Zelle klicken.</font></li>


Wenn sich die Zelle im Editier-Modus befindet, können Sie den Inhalt modifizieren, wie bei einem Normalen Text-Editierprogramm wie z.B. Microsoft Word.

Wenn sich die Zelle im Kommando-Modus befindet, können Sie das Notebook als ganzes editieren und z.B. ganze Zellen löschen, verschieben, kopieren, und einfügen aber Sie können nicht den Inhalt einzelner Zellen modifizieren. Verwenden Sie dafür das Menü, das sich rechts über der Zelle auftut sobald Sie eine Zelle auswählen.
