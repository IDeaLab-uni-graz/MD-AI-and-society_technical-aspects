```python
# relevante Module importieren
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

# Daten filtern und vorbereiten
daten_titanic = sns.load_dataset("titanic")
daten_titanic = daten_titanic.dropna(subset=["age", "fare", "survived"]).copy()
attribute = ["age", "fare"]
label = "survived"
x = daten_titanic[attribute]
y = daten_titanic[label]

# Klassifikator trainieren
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(x, y)

# Vorhersagen machen und in Tabelle speichern
vorhersagen = knn_classifier.predict(x)
daten_titanic["vorhersage"] = vorhersagen

# Vorhersagen visualisieren
sns.scatterplot(data=daten_titanic, x="age", y="fare", hue="vorhersage")
```
