## Übung 3
In der [Vorlesung](https://janalasser.at/lectures/MC_KI/VO3_4_ubueberwachtes_lernen_algorithmen/#/4/4/2) haben wir darüber gesprochen, dass der K Means Algorithmus nicht gut funktioniert, wenn die Cluster nicht "rund" geformt sind (keine konvexe Hülle haben). In diesem Fall kann dichtebasiertes Clustering eine Alternative darstellen.  

Im Folgenden wird ein künstlicher Beispieldatensatz erzeugt, der das Problem verdeutlichen soll:

```python
# erzeugen eines Beispieldatensatzes mit zwei "Monden", die ineinander greifen
from sklearn import datasets
monde = datasets.make_moons(n_samples=200, noise=0.05, random_state=42)[0]
monde = pd.DataFrame(monde, columns=["x", "y"])
sns.scatterplot(monde, x="x", y="y");
```

Wir versuchen, Cluster mit dem K Means Algorithmus zu finden. Offensichtlich gibt es in dem Datensatz zwei Gruppen von Beobachtungen, deswegen wählen wir `n_clusters=2`.

```python
# Clustering mit K Means Algorithmus und n_clusters=2
kmeans_monde_2 = KMeans(n_clusters=2, random_state=42)
kmeans_monde_2.fit(monde)
cluster_monde_2 = kmeans_monde_2.predict(monde)

# Erstellen eines DataFrames für die Visualisierung
plot_daten_monde = monde.copy()
plot_daten_monde["cluster_2"] = cluster_monde_2

# Visualisierung des Clusterings mit einer Punktwolke
sns.scatterplot(plot_daten_monde, x="x", y="y", hue="cluster_2");
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(monde, cluster_monde_2)
```

Der Silhouettenkoeffizient ist zwar nicht schlecht, das Clustering entspricht aber offensichtlich nicht der Struktur der Daten.
