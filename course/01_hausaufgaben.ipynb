## Schleifen

### Übung Schleifen 1

Versuchen Sie, den Wert der Variable `loops_per_loop` im nachfolgenden Programm langsam zu erhöhen. z.B. von 10 auf 100, 200, 300 usw. Zur Erinnerung: solange ein `*` neben der Zelle angezeigt wird, ist Python mit den Schleifen beschäftigt. Unterbrechen Sie den Prozess abbrechen, falls es zu lange dauern sollte. Diese Übung soll zeigen, wie schnell beim Verschachteln von Schleifen eine große Zahl von Schleifendurchläufen entstehen kann.

```python
durchlaufe_je_schleife = 10 # Anzahl der Durchläufe je Schleife

zaehler = 0
for i in range(durchlaufe_je_schleife):
    for j in range(durchlaufe_je_schleife):
        for k in range(durchlaufe_je_schleife):
            zaehler += 1
print(f"{zaehler} Schleifendurchläufe")
```

### Übung Schleifen 2
Im Kurs haben wir die `for` Schleife kennen gelernt, die über alle Elemente einer Sequenz (oder allgemeiner gesprochen: eines Iterables) iteriert. Darüber hinaus gibt es noch die `while`-Schleife, die bei jedem Schleifendurchlauf  eine Bedingung prüft und sich beendet, wenn diese Bedingung nicht mehr erfüllt ist.

Im folgenden Beispiel ist die Bedingung, dass die aktuelle Zahl nicht 10 ist. Sobald also die verglichene Zahl 10 ist, wird die `while`-Schleife beendet:

```python
i = 0
while i != 10:
    print(i)
    i += 1
```

Was würde passieren, wenn wir die letzte Zeile (i += 1) weglassen würden? Denken Sie dazu die ersten paar Schleifendurchläufe durch und schreiben Sie die Ausgabe von `print()` auf.

### Übung Listen 2 – mehrdimensionale Listen

Wir haben gesehen, dass eine Liste Werte mit beliebige Datentypen enthalten kann. Dazu zählen natürlich auch Listen. Wir können also eine Liste erzeugen, deren Elemente wiederum Listen sind. Wir haben dann eine Liste von Listen.

Stellen wir uns vor, wir messen drei Mal täglich die Temperatur und möchten diese Meßwerte speichern. Am ersten Tag haben wir diese 3 Messungen: `[17, 28, 24]`. Am zweiten Tage messen wird diese Werte: `[18, 31, 28]`. Am dritten Tag messen wir `[20, 35, 29]`. Wir haben also für jeden Tag eine Liste mit 3 Werten. Die einzelnen Tage (sprich: Listen) können wir wieder in einer Liste speichern:

```python
temperaturen = [
    [17, 28, 24],
    [18, 31, 28],
    [20, 35, 29]
]
```

Wir können uns diese Temperaturen als Tabelle vorstellen: Jede Zeile repräsentiert einen Tag, jede Spalte einen Meßzeitpunkt (z.B. 6:00, 12:00, 18:00). Wie wir auf die Messwerte eines bestimmten Tages zugreifen können, haben wir schon gelernt:

```python
temperaturen[1]
```

Da das so adressierte Element wieder eine Liste ist, können wir auch auf einzelne Element dieser Liste zugreifen. Den ersten Messwert des zweiten Tages können wir so auslesen:

```python
tag = temperaturen[1]     # Temperaturen des zweiten Tages als Liste
erste_temperatur = tag[0] # Erster Wert des Tages
erste_temperatur
```

Normalerweise schreibt man das kompakter so:

```python
temperaturen[1][0]
```

Der Wert im der ersten Klammernpaar beziehlt sich auch die Liste `temperaturen` (also das was wir oben als `tag` referenziert haben), der Wert im zweiten Klammernpaar auf die Liste, die wir über den ersten Index erhalten, also auf den ersten Eintrag des zweiten Eintrags der Liste `temperaturen`.

Damit sie mit diesem Konzept vertraut werden, lassen Sie sich folgende Temperaturen ausgeben:
* die zweite Temperatur des ersten Tages
* die erste Temperatur des ersten Tages
* die erste Temperatur des letzten Tages
* die letzte Temperatur des zweiten Tages
